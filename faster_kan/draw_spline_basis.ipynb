```python
import torch
from matplotlib import pyplot as plt
from efficient_kan import KANLinear
layer = KANLinear(28*28,  256 , 10)
# get b-spline basis (thx to efficient_kan)
x = torch.linspace(-1.6*2, 1.6, 28*28)
basis = layer.b_splines(x[None])
print(layer.grid)
print(basis[0, :, 0].max())
print(len(layer.grid[0]))
print(layer.grid[0][2])
print(layer.grid[0][-2])
print(64/layer.grid[0][-2]/4)
```

```python
# calculate radial basis functions with gaussian density kernel
def rbf(x, grid, h):
    return 1-torch.tanh((x[..., None] - grid)/h).pow(2)
rbf_basis = rbf(x, torch.linspace(layer.grid[0][2], 64/layer.grid[0][-2]/4, 64), 0.12)
rbf_basis.shape
fig = plt.figure(figsize=(8, 5))
ax = fig.subplots(1, 1)
for i in range(4):
    rbf_plt = ax.plot(x.numpy(), rbf_basis[:, i].numpy()*0.667 , color="red", label=None if i > 0 else "Reflectional SWitch RSWAF basis")
    bspline_plt = ax.plot(x.numpy(), basis[0, :, i].numpy(), "--", color="blue", label=None if i > 0 else "3-order B-spline basis")
ax.legend()
fig.savefig("img/compare_basis.png", dpi=100)
```

Conclusion: Reflectional SWitch Activation Function can approximate well 3-order B-Spline basis with linear scalings.
